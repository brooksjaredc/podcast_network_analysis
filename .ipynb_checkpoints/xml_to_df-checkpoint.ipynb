```python
tim_ferriss = xml_to_df_summ('tim-ferriss-show.xml')

def clean_tim_ferriss(tim_ferriss):

    ## Set Episode numbers
    
    #kevin_pereira['episode'] = 185 - kevin_pereira.index
    
    ## Convert duration time to seconds
    
    tim_ferriss['duration'] = [sum(x * int(t) for x, t in zip([1, 60, 3600], reversed(time.split(":")))) for time in tim_ferriss['duration']]
    
    ## Convert dates to datetime format
    
    dates_no_zeros =[]
    for d in tim_ferriss['date']:
        dates_no_zeros.append(re.sub('\s\-0000$', '', str(d), flags=re.IGNORECASE))
    tim_ferriss['date'] = [dt.datetime.strptime(d, '%a, %d %b %Y %H:%M:%S') for d in dates_no_zeros]
    
    tim_ferriss['guests'] = tim_ferriss.title.str.extract(r'(?:Episode \d+:\s)?([\w\s\,\.\']+)',expand=True)
    #tim_ferriss['guests'] = chris_hardwick['title']
    
#     guest_split(' #', chris_hardwick)
#     guest_split(' Returns', chris_hardwick)
#     guest_split(' returns', chris_hardwick)
#     guest_split(' LIVE', chris_hardwick)
#     guest_split(' Live', chris_hardwick)
#     guest_split('Live', chris_hardwick)
#     guest_split(' of', chris_hardwick)
    
#     for index, row in chris_hardwick.iterrows():
#         if(pd.notnull(row['guests'])):
#             if('w/ ' in row['guests']):
#                 split = row['guests'].split('/ ')[-1]
#                 chris_hardwick.at[index, 'guests'] = split


    
    
    ### Manual cleaning
    
    return tim_ferriss

tim_ferriss = clean_tim_ferriss(tim_ferriss)
    
tim_ferriss
```
